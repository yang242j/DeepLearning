```python
df_bbox = pd.read_csv('/kaggle/input/celeba-original-wild-images/list_bbox_celeba.csv')
df_bbox.head()
```

```python
def rowGen (imgFile, image, df):
    img_height = image.shape[0]
    img_width = image.shape[1]
    new_img_height = 218
    new_img_width = 178
    x_1 = int(df.x_1 * new_img_width / img_width)
    y_1 = int(df.y_1 * new_img_height / img_height)
    width = int(df.width * new_img_width / img_width)
    height = int(df.height * new_img_height / img_height)
    row = (imgFile, x_1, y_1, width, height)
    return row
```

```python
# open the file in the write mode
f = open('list_bbox_align_celeba.csv', 'w')

# create the csv writer
writer = csv.writer(f)

# write the first row
writer.writerow(('image_id','x_1','y_1','width','height'))

for imgFile in df_bbox.image_id:
    imgID = int(imgFile.replace('.jpg', ''))
    if(imgID <= 10000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 1/Part 1/'
    elif(imgID <= 20000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 2/Part 2/'
    elif(imgID <= 30000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 3/Part 3/'
    elif(imgID <= 40000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 4/Part 4/'
    elif(imgID <= 50000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 5/Part 5/'
    elif(imgID <= 60000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 6/Part 6/'
    elif(imgID <= 70000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 7/Part 7/'
    elif(imgID <= 80000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 8/Part 8/'
    elif(imgID <= 90000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 9/Part 9/'
    elif(imgID <= 100000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 10/Part 10/'
    elif(imgID <= 110000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 11/Part 11/'
    elif(imgID <= 120000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 12/Part 12/'
    elif(imgID <= 130000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 13/Part 13/'
    elif(imgID <= 140000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 14/Part 14/'
    elif(imgID <= 150000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 15/Part 15/'
    elif(imgID <= 160000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 16/Part 16/'
    elif(imgID <= 170000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 17/Part 17/'
    elif(imgID <= 180000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 18/Part 18/'
    elif(imgID <= 190000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 19/Part 19/'
    elif(imgID <= 200000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 20/Part 20/'
    elif(imgID <= 210000):
        part_path = '/kaggle/input/celeba-original-wild-images/Part 21/Part 21/'
    else:    
        print(imgFile, " Failed")
        
    try:
        image = cv2.imread(os.path.sep.join([part_path, imgFile]))
        df = df_bbox[df_bbox['image_id'] == imgFile]
        row = rowGen(imgFile, image, df)
        writer.writerow(row)
    except:
        print("image id: " + str(imgID))
        print("image file: " + str(imgFile))
        plt.imshow(image)
        plt.show()
        continue
    

# close the file
f.close()
```

```python
# orignial bbox
df = df_bbox[df_bbox['image_id'] == '160001.jpg']
x_1 = int(df.x_1)
y_1 = int(df.y_1)
width = int(df.width)
height = int(df.height)
print("x, y, w, h: ", x_1, y_1, width, height)

x_2 = x_1 + width
y_2 = y_1 + height
print("x, y, x, y: ", x_1, y_1, x_2, y_2)

# Draw bounding box
cv2.rectangle(image, (x_1, y_1), (x_2, y_2), (0, 255, 0), 2) # orignial bbox on original image

# Show image
plt.imshow(image)
plt.show()
```

```python
# resize image
image = cv2.imread('/kaggle/input/celeba-original-wild-images/Part 17/Part 17/160001.jpg')
new_img = cv2.resize(image, (178, 218))
print(new_img.shape)
plt.imshow(new_img)
plt.show()
```
